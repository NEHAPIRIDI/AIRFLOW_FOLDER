```python
import pandas as pd

# File paths
input_file = "Amazon.csv"
output_file = "processed_output.csv"

print("====== INPUT DATA NULL CHECK ======")

# Read only a sample for safety
input_df = pd.read_csv(input_file)

# Find rows with at least one NULL value
input_null_rows = input_df[input_df.isnull().any(axis=1)]

print(f"Total input rows: {len(input_df)}")
print(f"Rows with NULL values in input: {len(input_null_rows)}")

# Display first 5 null rows
print("\nSample NULL rows from INPUT:")
print(input_null_rows.head())

# -----------------------------------

print("\n====== OUTPUT DATA NULL CHECK ======")

output_df = pd.read_csv(output_file)

output_null_count = output_df.isnull().sum().sum()

print(f"Total output rows: {len(output_df)}")
print(f"Total NULL values in output: {output_null_count}")

if output_null_count == 0:
    print("✅ OUTPUT DATA IS CLEAN (NO NULL VALUES)")
else:
    print("❌ OUTPUT DATA STILL CONTAINS NULL VALUES")

```
